```python
processed_test_data=one_hot_test_data[:]
processed_test_data["Safety_Score"]=processed_test_data["Safety_Score"]/100
processed_test_data["Days_Since_Inspection"]=processed_test_data["Days_Since_Inspection"]/25
processed_test_data["Total_Safety_Complaints"]=processed_test_data["Total_Safety_Complaints"]/100
processed_test_data["Control_Metric"]=processed_test_data["Control_Metric"]/100
processed_test_data["Cabin_Temperature"]=processed_test_data["Cabin_Temperature"]/100
processed_test_data["Max_Elevation"]=processed_test_data["Max_Elevation"]/100000
processed_test_data["Violations"]=processed_test_data["Violations"]/5

processed_train_data=one_hot_train_data[:]
processed_train_data["Safety_Score"]=processed_train_data["Safety_Score"]/100
processed_train_data["Days_Since_Inspection"]=processed_train_data["Days_Since_Inspection"]/25
processed_train_data["Total_Safety_Complaints"]=processed_train_data["Total_Safety_Complaints"]/100
processed_train_data["Control_Metric"]=processed_train_data["Control_Metric"]/100
processed_train_data["Cabin_Temperature"]=processed_train_data["Cabin_Temperature"]/100
processed_train_data["Max_Elevation"]=processed_train_data["Max_Elevation"]/100000
processed_train_data["Violations"]=processed_train_data["Violations"]/5

#sample = np.random.choice(processed_train_data.index, size=int(len(processed_train_data)*0.4), replace=False)
processed_train_data, processed_valid_data = processed_train_data.iloc[:6000],processed_train_data.iloc[6000:]
processed_realtest_data=processed_valid_data.iloc[2000:]
processed_valid_data=processed_valid_data.iloc[:2000]
print(processed_realtest_data[:10])
print(processed_valid_data[:10])
```

```python
features_test=processed_test_data.drop(['Accident_ID'],axis=1)
test_Accident_ID=processed_test_data['Accident_ID']

features_train=processed_train_data.drop(['Highly_Fatal_And_Damaging','Minor_Damage_And_Injuries','Significant_Damage_And_Fatalities','Significant_Damage_And_Serious_Injuries','Accident_ID'],axis=1)
targets_train=processed_train_data[['Highly_Fatal_And_Damaging','Minor_Damage_And_Injuries','Significant_Damage_And_Fatalities','Significant_Damage_And_Serious_Injuries']]
features_valid=processed_valid_data.drop(['Highly_Fatal_And_Damaging','Minor_Damage_And_Injuries','Significant_Damage_And_Fatalities','Significant_Damage_And_Serious_Injuries','Accident_ID'],axis=1)
targets_valid=processed_valid_data[['Highly_Fatal_And_Damaging','Minor_Damage_And_Injuries','Significant_Damage_And_Fatalities','Significant_Damage_And_Serious_Injuries']]
features_realtest=processed_realtest_data.drop(['Highly_Fatal_And_Damaging','Minor_Damage_And_Injuries','Significant_Damage_And_Fatalities','Significant_Damage_And_Serious_Injuries','Accident_ID'],axis=1)
targets_realtest=processed_realtest_data[['Highly_Fatal_And_Damaging','Minor_Damage_And_Injuries','Significant_Damage_And_Fatalities','Significant_Damage_And_Serious_Injuries']]
train_dataset=[features_train,targets_train]
test_dataset=[features_test,test_Accident_ID]
valid_dataset=[features_valid,targets_valid]
realtest_dataset=[features_realtest,targets_realtest]
print(features_train[:1])
```

```python

import torch
from torch.utils.data import TensorDataset,DataLoader
```

```python
realtest_tens_dataset= TensorDataset(torch.FloatTensor(realtest_dataset[0].values),torch.LongTensor(realtest_dataset[1].values))
train_tens_dataset= TensorDataset(torch.FloatTensor(train_dataset[0].values),torch.FloatTensor(train_dataset[1].values))
valid_tens_dataset= TensorDataset(torch.FloatTensor(valid_dataset[0].values),torch.FloatTensor(valid_dataset[1].values))
test_tens_dataset = TensorDataset(torch.FloatTensor(test_dataset[0].values),torch.FloatTensor(test_dataset[1].values))


realtest_loader = DataLoader(realtest_tens_dataset, batch_size=20, shuffle=True)
train_loader = DataLoader(train_tens_dataset, batch_size=20, shuffle=True)
valid_loader = DataLoader(valid_tens_dataset, batch_size=20, shuffle=True)
test_loader = DataLoader(test_tens_dataset, batch_size=20)

dataiter = iter(realtest_loader)
features,targets= dataiter.next()
print(features[:10])
print(targets[:10])
print(len(test_tens_dataset))
```

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        hidden_1 = 64
        hidden_2 = 8
        self.fc1 = nn.Linear(16, hidden_1)
        self.fc2 = nn.Linear(hidden_1, hidden_2)
        self.fc3 = nn.Linear(hidden_2,4)    
        self.dropout1 = nn.Dropout(0.10)
        self.dropout2 = nn.Dropout(0.0)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout1(x)
        x = F.relu(self.fc2(x))
        x = self.dropout2(x)
        x = F.softmax(self.fc3(x), dim=1)
        return x

model = Net()
print(model(features))
```

```python
criterion = nn.MSELoss()

optimizer = torch.optim.Adamax(model.parameters(), lr=0.00001)
model.load_state_dict(torch.load('model6.pt'))
```

```python
# number of epochs to train the model
n_epochs =5000

# initialize tracker for minimum validation loss
valid_loss_min = np.Inf # set initial "min" to infinity

for epoch in range(n_epochs):
    # monitor training loss
    train_loss = 0.0
    valid_loss = 0.0
    
    ###################
    # train the model #
    ###################
    model.train() # prep model for training
    for data, target in train_loader:
        # clear the gradients of all optimized variables
        optimizer.zero_grad()
        # forward pass: compute predicted outputs by passing inputs to the model
        output = model(data)
        # calculate the loss
        loss = criterion(output, target)
        # backward pass: compute gradient of the loss with respect to model parameters
        loss.backward()
        # perform a single optimization step (parameter update)
        optimizer.step()
        # update running training loss
        train_loss += loss.item()*data.size(0)
        
    ######################    
    # validate the model #
    ######################
    model.eval() # prep model for evaluation
    for data, target in valid_loader:
        # forward pass: compute predicted outputs by passing inputs to the model
        output = model(data)
        # calculate the loss
        loss = criterion(output, target)
        # update running validation loss 
        valid_loss += loss.item()*data.size(0)
        
    # print training/validation statistics 
    # calculate average loss over an epoch
    train_loss = train_loss/len(train_loader.sampler)
    valid_loss = valid_loss/len(valid_loader.sampler)
    
    print('Epoch: {} \tTraining Loss: {:.6f} \tValidation Loss: {:.6f}'.format(
        epoch+1, 
        train_loss,
        valid_loss
        ))
    
    # save model if validation loss has decreased
    if valid_loss <= valid_loss_min:
        print('Validation loss decreased ({:.6f} --> {:.6f}).  Saving model ...'.format(
        valid_loss_min,
        valid_loss))
        torch.save(model.state_dict(), 'model6.pt')
        valid_loss_min = valid_loss
```

```python
classes=["Highly_Fatal_And_Damaging","Minor_Damage_And_Injuries","Significant_Damage_And_Fatalities","Significant_Damage_And_Serious_Injuries"]
switcher={"tensor([1, 0, 0, 0])":0,"tensor([0, 1, 0, 0])":1,"tensor([0, 0, 1, 0])":2,"tensor([0, 0, 0, 1])":3,"tensor([0, 0, 0, 0])":3}
```

```python
# initialize lists to monitor test loss and accuracy
test_loss = 0.0
class_correct = list(0. for i in range(4))
class_total = list(0. for i in range(4))

model.eval() # prep model for evaluation

for data, target in realtest_loader:
    # forward pass: compute predicted outputs by passing inputs to the model
    output = model(data)
    # calculate the loss
    loss = criterion(output, target.float())
    # update test loss 
    test_loss += loss.item()*data.size(0)
    # convert output probabilities to predicted class
    pred =output.round().long()
    # compare predictions to true label
    correct = np.squeeze(pred.eq(target.data.view_as(pred)))
    # calculate test accuracy for each object class
    for i in range(len(target)):
        label = target.data[i]
        label=switcher[str(label)]
        class_correct[label] += 1 if sum(correct[i])%4==0 else 0
        class_total[label] += 1

# calculate and print avg test loss
test_loss = test_loss/len(realtest_loader.sampler)
print('Test Loss: {:.6f}\n'.format(test_loss))

for i in range(4):
    if class_total[i] > 0:
        print('Test Accuracy of %5s: %2d%% (%2d/%2d)' % (
            classes[i], 100 * class_correct[i] / class_total[i],
            np.sum(class_correct[i]), np.sum(class_total[i])))
    else:
        print('Test Accuracy of %5s: N/A (no training examples)' % (classes[i]))

print('\nTest Accuracy (Overall): %2d%% (%2d/%2d)' % (
    100. * np.sum(class_correct) / np.sum(class_total),
    np.sum(class_correct), np.sum(class_total)))
```
